## Matplotlib

```python
#First we create some example data. Then we will create a few plots
x = np.arange(0, 2*np.pi, .01)
y = np.sin(x)

plt.figure(figsize=(8,8))
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('$Y_\mu$') # We can also use latex notation
plt.title('My first matplotlib plot')
```

We can of course do more complex plots. 

```python
x = np.random.randn(10000)
y = np.random.randn(10000)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.scatter(x, y)

plt.subplot(1,2,2)
plt.hist2d(x, y, bins=100, range=((-3,3),(-3,3)));
```

We can even plot in 3D. 

```python
from mpl_toolkits.mplot3d import axes3d


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
X, Y, Z = axes3d.get_test_data(0.05)

# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```

## Numpy
The second library that most of the scientific python ecosystem is built upon is numpy. While we don't have time to completely cover it, we can show a few of the functions that we will use most. 

In numpy we will primarily use arrays `np.array`. These function somewhat similar to matlab arrays, except most operations that work on these arrays will also be inside numpy. Let's step through some common uses. 

```python
x = np.arange(0, 2*np.pi, .01)  # Create an array increasing from 0 to 100 in increments of 1
```

```python
y = np.sin(x)  # Calculate the sin of x and store it in y
```

```python
# Quick exercvise: Why doesn't this work? 
y = sin(x)
```

We can access individual elements of an array , as well as ranges. 

```python
print(y[100])
```

```python
print(y[100:105])
```

Similarly we can setup and change value this way. 

```python
y = np.arange(0, 10, 1.0)
print(y)
y[5] = 100
print(y)
y[5:7] = [1,2]
print(y)
```

Arrays have type information as well, though we will usually be working with floating point numbers. 

```python
y.dtype
```

We will also be using multidimensional arrays (referred to as tensors in some other libraries). 

```python
y = np.random.randn(3)
t = np.outer(y, y.T)
print(f't is {t}\n')
print(f'The first element of t is {t[0]}')
print(f't[0,0] is {t[0,0]}')
```
